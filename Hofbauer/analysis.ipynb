# Analysis

## Imports

```python
import matplotlib.pyplot as plt
import torch
```

## Model 1

```python
model_one = torch.load('data/model_checkpoints/model_one.pt')
y_test = model_one['y_test']
model_1_predict = model_one['model_1_predict']
# Assuming model_1_predict is the predicted values and y_test is the actual values
plt.scatter(y_test, model_1_predict, color='blue')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], linestyle='--', color='red', linewidth=2)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted Values')
plt.show()
```
