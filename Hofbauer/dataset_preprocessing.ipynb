```python
# Categorical data
categorical_columns = ['gender', 'race', 'age_range', 'has_tiktok', 'remembers_disco', 'uses_skincare']

# Plot histograms
utils.plot_categorical_histograms(df,categorical_columns)
```

```python
#Counts for each catergorical column
for i in range(len(categorical_columns)):
    print(f"\nColumn: {categorical_columns[i]}")
    counts = df[categorical_columns[i]].value_counts()
    for label, count in counts.items():
        print(f"Label: '{label}' | Frequency: {count}")
```

 #### Dropping Features

```python
#Numeric
#None
#Catergoical
df = df.drop('race', axis=1)
df = df.drop('age_range', axis=1)

print(df.dtypes)
```

#### Normalization

```python
#Numeric
df['num_haircuts_life'] = utils.standardize_numeric(df['num_haircuts_life'], False)
df['max_annual_earnings'] = utils.standardize_numeric(df['max_annual_earnings'], False)

#Catergorical
categorical_columns = ['gender','has_tiktok', 'remembers_disco', 'uses_skincare']
df = utils.one_hot_encode_columns(df, categorical_columns)
```

### Conversion

#### Defining Input and Outputs

```python
features = ['num_haircuts_life','max_annual_earnings','gender_male','has_tiktok_yes', 'remembers_disco_yes', 'uses_skincare_yes']
target = ['age']

#new dataframes for the inputs and outputs
train_df = df[features + target]
```

#### Train Test Split

```python
# Split train and val 
x_train, x_val, y_train, y_val = train_test_split(train_df[features], train_df[target], train_size=0.6, random_state=27)

# Split again to get a test set
x_val, x_test, y_val, y_test = train_test_split(x_val, y_val, train_size=0.5, random_state=27)

print("x train: ",x_train.shape, "y train:", y_train.shape)
print("x val: ",x_val.shape, "y train:", y_val.shape)
print("x test: ",x_test.shape, "y test:", y_test.shape)
```

##### Saving Processed Data

```python
# dataframes -> numpy arrays -> tensors
print(x_train)
x_train, x_val, x_test = torch.Tensor(x_train.to_numpy()), torch.Tensor(x_val.to_numpy()), torch.Tensor(x_test.to_numpy())
y_train, y_val, y_test = torch.Tensor(y_train.to_numpy()), torch.Tensor(y_val.to_numpy()), torch.Tensor(y_test.to_numpy())

# store it in a dict that we can save out as a single file
data_dict = {'x_train':x_train, 'x_val':x_val, 'x_test':x_test, 'y_train':y_train, 'y_val':y_val, 'y_test':y_test}

# save it to local data directory
torch.save(data_dict, 'data/UTKFace_processed.pt')
```

### Train Test Split

```python
features = ['filename']
target = ['age']

#new dataframes for the inputs and outputs
train_df = df[features + target]
```

```python
#Split the filenames and associated ages
filenames = train_df['filename'].tolist()
ages = train_df['age'].tolist()

# Split train, val, and test indexes
train_indexes, val_test_indexes = train_test_split(range(len(filenames)), train_size=0.6, random_state=27)
val_indexes, test_indexes = train_test_split(val_test_indexes, train_size=0.5, random_state=27)

# Create subsets based on the splits
train_filenames = [filenames[i] for i in train_indexes]
val_filenames = [filenames[i] for i in val_indexes]
test_filenames = [filenames[i] for i in test_indexes]

train_ages = [ages[i] for i in train_indexes]
val_ages = [ages[i] for i in val_indexes]
test_ages = [ages[i] for i in test_indexes]

# Store the split lists in a dictionary
split_data = {
    'x_train': train_filenames,
    'y_train': train_ages,
    'x_val': val_filenames,
    'y_val': val_ages,
    'x_test': test_filenames,
    'y_test': test_ages,
}

# Save the split lists to a file
torch.save(split_data, 'data/filename_splits.pt')
```
